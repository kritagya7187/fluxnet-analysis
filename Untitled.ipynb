```python
import os
import zipfile
import pandas as pd
import chardet
import shutil

# ============================================================
# CONFIGURATION
# ============================================================
DATA_DIR = r"C:\Users\BRAHMA LAB\Downlaod\mpi_fail\fluxnet"
OUTPUT_DIR = r"C:\Users\BRAHMA LAB\fluxnet_processed"
os.makedirs(OUTPUT_DIR, exist_ok=True)

# Hydraulic classification (same as before)
HYDRAULIC_CLASS = {
    'US-Ha1': 'anisohydric', 'US-MMS': 'anisohydric', 'US-WCr': 'anisohydric',
    'FR-Pue': 'anisohydric', 'BR-Ma2': 'anisohydric', 'CA-Cbo': 'anisohydric',
    'CA-Gro': 'anisohydric', 'CA-TPD': 'anisohydric', 'BR-Ji3': 'anisohydric',
    'BR-Sa1': 'anisohydric', 'CL-SDF': 'anisohydric',
    'IT-Ren': 'isohydric', 'DE-Tha': 'isohydric', 'CA-Ca1': 'isohydric',
    'CA-Ca2': 'isohydric', 'CA-Ca3': 'isohydric', 'CA-LP1': 'isohydric',
    'CA-TP1': 'isohydric', 'CA-TP3': 'isohydric', 'CA-TP4': 'isohydric',
}

def detect_encoding(file_path):
    with open(file_path, 'rb') as f:
        raw = f.read(100000)
        result = chardet.detect(raw)
        return result['encoding']

def process_site(zip_path, site_id):
    extract_dir = os.path.join(OUTPUT_DIR, f"temp_{site_id}")
    os.makedirs(extract_dir, exist_ok=True)
    
    with zipfile.ZipFile(zip_path, 'r') as zf:
        zf.extractall(extract_dir)
    
    # Find all CSV files (search recursively)
    csv_files = []
    for root, dirs, files in os.walk(extract_dir):
        for f in files:
            if f.endswith('.csv'):
                csv_files.append(os.path.join(root, f))
    
    # Pick the half‑hourly FLUXMET file: contains 'FLUXMET' and ('_HH_' or '_HR_'), and not 'BIF' or 'VARINFO'
    hr_file = None
    for fpath in csv_files:
        fname = os.path.basename(fpath)
        if 'FLUXMET' in fname and ('_HH_' in fname or '_HR_' in fname):
            if 'BIF' not in fname and 'VARINFO' not in fname:
                hr_file = fpath
                break
    
    if hr_file is None:
        raise FileNotFoundError(f"No half-hourly FLUXMET file for {site_id}")
    
    # Detect encoding
    enc = detect_encoding(hr_file)
    print(f"  Encoding: {enc}")
    
    # Read CSV
    df = pd.read_csv(hr_file, na_values=-9999, encoding=enc)
    
    # Parse timestamp
    df['datetime'] = pd.to_datetime(df['TIMESTAMP_START'], format='%Y%m%d%H%M', errors='coerce')
    df = df.dropna(subset=['datetime'])
    
    # QC: keep only good quality for NEE and LE (if QC columns exist)
    if 'NEE_VUT_REF_QC' in df.columns and 'LE_F_MDS_QC' in df.columns:
        df = df[(df['NEE_VUT_REF_QC'] < 0.75) & (df['LE_F_MDS_QC'] < 0.75)]
    
    # GPP column
    if 'GPP_NT_VUT_REF' in df.columns:
        gpp_col = 'GPP_NT_VUT_REF'
    elif 'GPP_DT_VUT_REF' in df.columns:
        gpp_col = 'GPP_DT_VUT_REF'
    else:
        raise KeyError(f"No GPP column for {site_id}")
    
    # LE column
    le_col = 'LE_F_MDS'
    if le_col not in df.columns:
        raise KeyError(f"No LE_F_MDS column for {site_id}")
    
    # Daily aggregation
    daily = df.resample('D', on='datetime').agg({
        gpp_col: 'mean',
        le_col: 'mean',
    }).dropna()
    
    daily['ET_mm'] = daily[le_col] * 0.0353
    daily = daily.rename(columns={gpp_col: 'GPP_gC_m2_day'})
    daily = daily[['GPP_gC_m2_day', 'ET_mm']]
    daily['site_id'] = site_id
    daily['hydraulic'] = HYDRAULIC_CLASS.get(site_id, 'unknown')
    
    # Cleanup
    shutil.rmtree(extract_dir)
    return daily

# ============================================================
# MAIN LOOP
# ============================================================
all_daily = []
zip_files = [f for f in os.listdir(DATA_DIR) if f.endswith('.zip')]

for zip_file in zip_files:
    # Extract site ID
    if zip_file.startswith('AMF_'):
        site_id = zip_file.split('_')[1]
    elif zip_file.startswith('ICOS_'):
        site_id = zip_file.split('_')[1]
    else:
        site_id = zip_file.split('_')[0]
    
    print(f"Processing {site_id}...")
    zip_path = os.path.join(DATA_DIR, zip_file)
    try:
        daily_df = process_site(zip_path, site_id)
        all_daily.append(daily_df)
        print(f"  -> {len(daily_df)} days")
    except Exception as e:
        print(f"  ERROR: {e}")

if all_daily:
    combined = pd.concat(all_daily)
    out_path = os.path.join(OUTPUT_DIR, "all_sites_daily_GPP_ET.csv")
    combined.to_csv(out_path)
    print(f"\n✓ Saved {len(combined)} records to {out_path}")
    print("\nRecords per site:")
    print(combined.groupby(['site_id', 'hydraulic']).size())
else:
    print("No data processed.")
```

```python
import os, zipfile, pandas as pd, numpy as np
import matplotlib.pyplot as plt

DATA_DIR = r"C:\Users\BRAHMA LAB\Downlaod\mpi_fail\fluxnet"
OUTPUT_DIR = r"C:\Users\BRAHMA LAB\fluxnet_processed"
os.makedirs(OUTPUT_DIR, exist_ok=True)

hydraulic = {
    'US-Ha1':'ani','US-MMS':'ani','US-WCr':'ani','FR-Pue':'ani','BR-Ma2':'ani',
    'CA-Cbo':'ani','CA-Gro':'ani','CA-TPD':'ani','BR-Ji3':'ani','BR-Sa1':'ani',
    'CL-SDF':'ani','IT-Ren':'iso','DE-Tha':'iso','CA-Ca1':'iso','CA-Ca2':'iso',
    'CA-Ca3':'iso','CA-LP1':'iso','CA-TP1':'iso','CA-TP3':'iso','CA-TP4':'iso'
}

def extract_site(zip_path):
    site = os.path.basename(zip_path).split('_')[1]
    with zipfile.ZipFile(zip_path, 'r') as zf:
        candidates = [f for f in zf.namelist() if ('HH' in f or 'HR' in f) and 'FLUXMET' in f and 'BIF' not in f]
        if not candidates:
            return None
        with zf.open(candidates[0]) as f:
            df = pd.read_csv(f, encoding='latin1', na_values=-9999)
    # Timestamp
    if 'TIMESTAMP_START' not in df:
        return None
    df['datetime'] = pd.to_datetime(df['TIMESTAMP_START'], format='%Y%m%d%H%M', errors='coerce')
    df = df.dropna(subset=['datetime'])
    # QC
    if 'NEE_VUT_REF_QC' in df:
        df = df[df['NEE_VUT_REF_QC'] < 0.75]
    if 'LE_F_MDS_QC' in df:
        df = df[df['LE_F_MDS_QC'] < 0.75]
    # Find GPP column
    gpp_col = None
    for cand in ['GPP_NT_VUT_REF', 'GPP_DT_VUT_REF', 'GPP']:
        if cand in df.columns:
            gpp_col = cand
            break
    if gpp_col is None:
        print(f"  No GPP column for {site}")
        return None
    # Find LE column
    le_col = None
    for cand in ['LE_F_MDS', 'LE_CORR', 'LE']:
        if cand in df.columns:
            le_col = cand
            break
    if le_col is None:
        print(f"  No LE column for {site}")
        return None
    # VPD optional
    vpd_col = 'VPD_F_MDS' if 'VPD_F_MDS' in df.columns else None
    # Resample to daily
    daily = df.resample('D', on='datetime').agg({gpp_col: 'mean', le_col: 'mean'})
    if vpd_col:
        daily['VPD'] = df.resample('D', on='datetime')[vpd_col].mean()
    daily['ET_mm'] = daily[le_col] * 0.0353
    daily = daily.rename(columns={gpp_col: 'GPP'})[['GPP', 'ET_mm'] + (['VPD'] if vpd_col else [])]
    daily['site'] = site
    daily['group'] = hydraulic.get(site, 'unknown')
    return daily

all_data = []
for zip_file in os.listdir(DATA_DIR):
    if not zip_file.endswith('.zip'): continue
    path = os.path.join(DATA_DIR, zip_file)
    print(f"Processing {zip_file.split('_')[1]}...")
    df = extract_site(path)
    if df is not None and len(df) > 100:
        all_data.append(df)
        print(f"  -> {len(df)} days")
    else:
        print(f"  -> skipped (no data)")

if not all_data:
    raise ValueError("No data extracted. Check file contents.")

combined = pd.concat(all_data)
combined.to_csv(os.path.join(OUTPUT_DIR, "daily_GPP_ET_VPD.csv"), index=False)
print(f"\nSaved {len(combined)} daily records")

# ------------------- Drought and asymmetry -------------------
def find_droughts(series, threshold=90, min_days=5):
    high = series > np.percentile(series.dropna(), threshold)
    runs = (high != high.shift()).cumsum()
    droughts = []
    for run, group in high.groupby(runs):
        if group.iloc[0] and len(group) >= min_days:
            droughts.append((group.index[0], group.index[-1]))
    return droughts

results = []
for site, site_df in combined.groupby('site'):
    if 'VPD' not in site_df.columns or site_df['VPD'].isna().all():
        print(f"Skipping {site}: no VPD data")
        continue
    droughts = find_droughts(site_df['VPD'], threshold=90, min_days=5)
    for start, end in droughts:
        pre = site_df[(site_df.index >= start - pd.Timedelta(days=30)) & (site_df.index < start)]
        post = site_df[(site_df.index > end) & (site_df.index <= end + pd.Timedelta(days=30))]
        if len(pre) < 10 or len(post) < 10:
            continue
        gpp_pre = pre['GPP'].mean()
        et_pre = pre['ET_mm'].mean()
        if gpp_pre <= 0 or et_pre <= 0:
            continue
        gpp_rec = post['GPP'].mean()
        et_rec = post['ET_mm'].mean()
        asym = (et_rec / et_pre) - (gpp_rec / gpp_pre)
        results.append({
            'site': site,
            'group': site_df['group'].iloc[0],
            'asymmetry': asym,
            'gpp_ratio': gpp_rec/gpp_pre,
            'et_ratio': et_rec/et_pre
        })

res_df = pd.DataFrame(results)
print("\nAsymmetry results:")
print(res_df.groupby('group')['asymmetry'].agg(['mean','std','count']))

# Plot
fig, ax = plt.subplots(figsize=(6,4))
colors = {'iso':'blue', 'ani':'red'}
for g in res_df['group'].unique():
    data = res_df[res_df['group']==g]['asymmetry']
    ax.scatter([g]*len(data), data, color=colors[g], alpha=0.6, label=f'{g} (n={len(data)})')
    ax.axhline(data.mean(), color=colors[g], linestyle='--')
ax.axhline(0, color='black', linestyle=':')
ax.set_ylabel('Asymmetry (ET/GPP recovery ratio difference)')
ax.set_title('Post‑drought asymmetry: anisohydric vs isohydric')
ax.legend()
plt.tight_layout()
plt.savefig(os.path.join(OUTPUT_DIR, "asymmetry_plot.png"), dpi=150)
plt.show()
print(f"\nPlot saved to {OUTPUT_DIR}\\asymmetry_plot.png")
```

```python
import os
import zipfile
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

DATA_DIR = r"C:\Users\BRAHMA LAB\Downlaod\mpi_fail\fluxnet"
OUTPUT_DIR = r"C:\Users\BRAHMA LAB\fluxnet_processed"
os.makedirs(OUTPUT_DIR, exist_ok=True)

# Hydraulic classification
hydraulic = {
    'US-Ha1':'ani', 'US-MMS':'ani', 'US-WCr':'ani', 'FR-Pue':'ani', 'BR-Ma2':'ani',
    'CA-Cbo':'ani', 'CA-Gro':'ani', 'CA-TPD':'ani', 'BR-Ji3':'ani', 'BR-Sa1':'ani',
    'CL-SDF':'ani', 'IT-Ren':'iso', 'DE-Tha':'iso', 'CA-Ca1':'iso', 'CA-Ca2':'iso',
    'CA-Ca3':'iso', 'CA-LP1':'iso', 'CA-TP1':'iso', 'CA-TP3':'iso', 'CA-TP4':'iso'
}

def extract_site(zip_path):
    site = os.path.basename(zip_path).split('_')[1]
    with zipfile.ZipFile(zip_path, 'r') as zf:
        # Find half-hourly FLUXMET file (_HH_ or _HR_)
        candidates = [f for f in zf.namelist() if ('HH' in f or 'HR' in f) and 'FLUXMET' in f and 'BIF' not in f]
        if not candidates:
            return None
        with zf.open(candidates[0]) as f:
            df = pd.read_csv(f, encoding='latin1', na_values=-9999)
    # Timestamp
    if 'TIMESTAMP_START' not in df:
        return None
    df['datetime'] = pd.to_datetime(df['TIMESTAMP_START'], format='%Y%m%d%H%M', errors='coerce')
    df = df.dropna(subset=['datetime'])
    # QC
    if 'NEE_VUT_REF_QC' in df:
        df = df[df['NEE_VUT_REF_QC'] < 0.75]
    if 'LE_F_MDS_QC' in df:
        df = df[df['LE_F_MDS_QC'] < 0.75]
    # GPP
    gpp_col = 'GPP_NT_VUT_REF' if 'GPP_NT_VUT_REF' in df else 'GPP_DT_VUT_REF'
    if gpp_col not in df:
        return None
    # LE
    le_col = 'LE_F_MDS' if 'LE_F_MDS' in df else 'LE'
    if le_col not in df:
        return None
    # VPD (optional)
    vpd_col = 'VPD_F_MDS' if 'VPD_F_MDS' in df else None
    # SWC (soil moisture, optional)
    swc_col = None
    for cand in ['SWC_F_MDS_1', 'SWC_F_MDS', 'SWC']:
        if cand in df.columns:
            swc_col = cand
            break
    # Daily aggregation
    daily = df.resample('D', on='datetime').agg({
        gpp_col: 'mean',
        le_col: 'mean'
    })
    if vpd_col:
        daily['VPD'] = df.resample('D', on='datetime')[vpd_col].mean()
    if swc_col:
        daily['SWC'] = df.resample('D', on='datetime')[swc_col].mean()
    daily['ET_mm'] = daily[le_col] * 0.0353
    daily['GPP'] = daily[gpp_col]
    daily = daily[['GPP', 'ET_mm'] + ([c for c in ['VPD','SWC'] if c in daily.columns])]
    daily['site'] = site
    daily['group'] = hydraulic.get(site, 'unknown')
    daily.reset_index(inplace=True)   # make datetime a column
    return daily

# Extract all sites
all_data = []
for zip_file in os.listdir(DATA_DIR):
    if not zip_file.endswith('.zip'):
        continue
    path = os.path.join(DATA_DIR, zip_file)
    print(f"Extracting {zip_file.split('_')[1]}...")
    df = extract_site(path)
    if df is not None and len(df) > 100:
        all_data.append(df)
        print(f"  -> {len(df)} days")
    else:
        print(f"  -> failed or too short")

if not all_data:
    raise ValueError("No data extracted")

combined = pd.concat(all_data, ignore_index=True)
combined.to_csv(os.path.join(OUTPUT_DIR, "daily_full.csv"), index=False)
print(f"\nSaved {len(combined)} records to daily_full.csv")

# ------------------- Drought and asymmetry -------------------
def find_droughts_swc(swc_series, threshold_percentile=10, min_days=5):
    thresh = np.percentile(swc_series.dropna(), threshold_percentile)
    low = swc_series < thresh
    runs = (low != low.shift()).cumsum()
    droughts = []
    for run, group in low.groupby(runs):
        if group.iloc[0] and len(group) >= min_days:
            droughts.append((group.index[0], group.index[-1]))
    return droughts

def estimate_transpiration(et_mm, evap_fraction=0.4):
    return et_mm * (1 - evap_fraction)

results = []
for site, site_df in combined.groupby('site'):
    site_df = site_df.sort_values('datetime').set_index('datetime')
    if 'SWC' not in site_df.columns or site_df['SWC'].isna().all():
        print(f"Skipping {site}: no SWC")
        continue
    site_df['T_mm'] = estimate_transpiration(site_df['ET_mm'])
    droughts = find_droughts_swc(site_df['SWC'], threshold_percentile=10, min_days=5)
    for start, end in droughts:
        pre = site_df[(site_df.index >= start - pd.Timedelta(days=30)) & (site_df.index < start)]
        post = site_df[(site_df.index > end) & (site_df.index <= end + pd.Timedelta(days=90))]
        if len(pre) < 10 or len(post) < 10:
            continue
        gpp_pre = pre['GPP'].mean()
        et_pre = pre['ET_mm'].mean()
        t_pre = pre['T_mm'].mean()
        if gpp_pre <= 0 or et_pre <= 0 or t_pre <= 0:
            continue
        gpp_rec = post['GPP'].mean()
        et_rec = post['ET_mm'].mean()
        t_rec = post['T_mm'].mean()
        asym_et = (et_rec / et_pre) - (gpp_rec / gpp_pre)
        asym_t  = (t_rec / t_pre) - (gpp_rec / gpp_pre)
        results.append({
            'site': site,
            'group': site_df['group'].iloc[0],
            'asymmetry_ET_total': asym_et,
            'asymmetry_Transpiration': asym_t,
        })

res_df = pd.DataFrame(results)
print("\n=== Asymmetry Results (90‑day recovery, SWC drought) ===")
print("Total ET asymmetry:")
print(res_df.groupby('group')['asymmetry_ET_total'].agg(['mean','std','count']))
print("\nTranspiration asymmetry:")
print(res_df.groupby('group')['asymmetry_Transpiration'].agg(['mean','std','count']))

# Plot
fig, axes = plt.subplots(1,2, figsize=(10,4))
colors = {'iso':'blue', 'ani':'red'}
metrics = ['asymmetry_ET_total', 'asymmetry_Transpiration']
titles = ['Total ET Asymmetry', 'Transpiration Asymmetry']
for ax, metric, title in zip(axes, metrics, titles):
    for g in res_df['group'].unique():
        data = res_df[res_df['group']==g][metric].dropna()
        if len(data) == 0: continue
        ax.scatter([g]*len(data), data, color=colors[g], alpha=0.6, label=f'{g} (n={len(data)})')
        ax.axhline(data.mean(), color=colors[g], linestyle='--')
    ax.axhline(0, color='black', linestyle=':')
    ax.set_ylabel('Asymmetry (recovery ratio diff)')
    ax.set_title(title)
    ax.legend()
plt.tight_layout()
plt.savefig(os.path.join(OUTPUT_DIR, "asymmetry_swc_90d.png"), dpi=150)
plt.show()
print(f"\nPlot saved to {OUTPUT_DIR}\\asymmetry_swc_90d.png")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
import os

# ============================================================
# CONFIGURATION
# ============================================================
INPUT_CSV = r"C:\Users\BRAHMA LAB\fluxnet_processed\daily_full.csv"
OUTPUT_DIR = r"C:\Users\BRAHMA LAB\fluxnet_processed"
os.makedirs(OUTPUT_DIR, exist_ok=True)

DROUGHT_PERCENTILE = 90      # VPD above this percentile = drought
MIN_DROUGHT_DAYS = 5
PRE_DAYS = 30
POST_DAYS = 90
CONTROL_N_SAMPLES = 100      # number of random non-drought windows per site

# ============================================================
# LOAD DATA
# ============================================================
df = pd.read_csv(INPUT_CSV, parse_dates=['datetime'])
df = df.sort_values(['site', 'datetime'])

# Keep only sites with VPD data
df = df.dropna(subset=['VPD'])
sites = df['site'].unique()
print(f"Analyzing {len(sites)} sites with VPD data.")

# ============================================================
# FUNCTION TO FIND DROUGHTS (using VPD)
# ============================================================
def find_droughts(vpd_series, dates, percentile=90, min_days=5, merge_gap_days=30):
    """
    vpd_series: pandas Series with datetime index (VPD values)
    dates: datetime column (not used directly, but kept for signature consistency)
    """
    thresh = np.percentile(vpd_series.dropna(), percentile)
    high = (vpd_series > thresh).rename('high')   # give the boolean series a name
    # Convert to DataFrame for easy manipulation
    high_df = high.reset_index()
    high_df['date'] = pd.to_datetime(high_df['datetime'])
    high_df = high_df.set_index('date')
    high_df['gap'] = high_df.index.to_series().diff().dt.days > 1
    high_df['block'] = (high_df['gap'] | (high_df['high'] != high_df['high'].shift())).cumsum()
    droughts = []
    for block, group in high_df.groupby('block'):
        if group['high'].iloc[0] and len(group) >= min_days:
            start = group.index.min()
            end = group.index.max()
            droughts.append((start, end))
    # Merge close droughts
    if merge_gap_days > 0 and len(droughts) > 1:
        merged = []
        cur_start, cur_end = droughts[0]
        for s, e in droughts[1:]:
            if (s - cur_end).days <= merge_gap_days:
                cur_end = max(cur_end, e)
            else:
                merged.append((cur_start, cur_end))
                cur_start, cur_end = s, e
        merged.append((cur_start, cur_end))
        droughts = merged
    return droughts

# ============================================================
# FUNCTION TO COMPUTE DECOUPLING (absolute asymmetry)
# ============================================================
def compute_decoupling_safe(site_df, start, end, pre_days, post_days, 
                            min_gpp_pre=0.5, min_et_pre=0.2, max_decoupling=5.0):
    pre_mask = (site_df['datetime'] >= start - pd.Timedelta(days=pre_days)) & (site_df['datetime'] < start)
    post_mask = (site_df['datetime'] > end) & (site_df['datetime'] <= end + pd.Timedelta(days=post_days))
    pre = site_df[pre_mask]
    post = site_df[post_mask]
    if len(pre) < 10 or len(post) < 10:
        return np.nan
    gpp_pre = pre['GPP'].mean()
    et_pre = pre['ET_mm'].mean()
    if gpp_pre < min_gpp_pre or et_pre < min_et_pre:
        return np.nan
    gpp_rec = post['GPP'].mean()
    et_rec = post['ET_mm'].mean()
    asym = (et_rec / et_pre) - (gpp_rec / gpp_pre)
    dec = np.abs(asym)
    return min(dec, max_decoupling)   # cap outliers

# ============================================================
# COLLECT DROUGHT DECOUPLING
# ============================================================
drought_decoupling = []

for site in sites:
    site_df = df[df['site'] == site].sort_values('datetime').copy()
    # Use VPD as drought metric
    site_df = site_df.set_index('datetime')
    vpd_series = site_df['VPD']
    site_df = site_df.reset_index()  # keep datetime as column
    droughts = find_droughts(vpd_series, site_df['datetime'],
                             percentile=DROUGHT_PERCENTILE,
                             min_days=MIN_DROUGHT_DAYS,
                             merge_gap_days=30)
    for start, end in droughts:
        dec = compute_decoupling(site_df, start, end, PRE_DAYS, POST_DAYS)
        if not np.isnan(dec):
            drought_decoupling.append(dec)
    print(f"{site}: {len(droughts)} droughts")

# ============================================================
# COLLECT CONTROL (NON-DROUGHT) DECOUPLING
# ============================================================
control_decoupling = []
rng = np.random.default_rng(42)

for site in sites:
    site_df = df[df['site'] == site].sort_values('datetime').copy()
    site_df_idx = site_df.set_index('datetime')
    vpd_series = site_df_idx['VPD']
    droughts = find_droughts(vpd_series, site_df['datetime'],
                             percentile=DROUGHT_PERCENTILE,
                             min_days=MIN_DROUGHT_DAYS,
                             merge_gap_days=30)
    # Build set of drought dates
    drought_dates = set()
    for s, e in droughts:
        drought_dates.update(pd.date_range(s, e))
    # Get all dates not in drought
    all_dates = site_df['datetime'].values
    non_drought = [d for d in all_dates if d not in drought_dates]
    if len(non_drought) < PRE_DAYS + POST_DAYS + 10:
        continue
    max_start = len(non_drought) - (PRE_DAYS + POST_DAYS)
    if max_start < 1:
        continue
    for _ in range(CONTROL_N_SAMPLES):
        # pick a random start index from non_drought list
        idx = rng.integers(0, max_start)
        start = non_drought[idx]
        end = start + pd.Timedelta(days=POST_DAYS - 1)
        # Ensure the whole window (including pre period) is drought-free
        pre_start = start - pd.Timedelta(days=PRE_DAYS)
        window_dates = set(pd.date_range(pre_start, end))
        if window_dates.intersection(drought_dates):
            continue
        dec = compute_decoupling(site_df, start, end, PRE_DAYS, POST_DAYS)
        if not np.isnan(dec):
            control_decoupling.append(dec)

# ============================================================
# STATISTICAL COMPARISON
# ============================================================
d_arr = np.array(drought_decoupling)
c_arr = np.array(control_decoupling)

print(f"\nDrought events: {len(d_arr)}, mean decoupling = {np.mean(d_arr):.4f} ± {np.std(d_arr):.4f}")
print(f"Control windows: {len(c_arr)}, mean decoupling = {np.mean(c_arr):.4f} ± {np.std(c_arr):.4f}")

# Mann-Whitney U test (one‑tailed: drought > control)
u_stat, p_val = stats.mannwhitneyu(d_arr, c_arr, alternative='greater')
print(f"Mann-Whitney U: p = {p_val:.6f}")

# Bootstrap difference of means
n_boot = 1000
diffs = []
for _ in range(n_boot):
    boot_d = rng.choice(d_arr, size=len(d_arr), replace=True)
    boot_c = rng.choice(c_arr, size=len(c_arr), replace=True)
    diffs.append(np.mean(boot_d) - np.mean(boot_c))
ci_low, ci_high = np.percentile(diffs, [2.5, 97.5])
print(f"Bootstrap 95% CI for difference (drought - control): [{ci_low:.4f}, {ci_high:.4f}]")

# ============================================================
# PLOT
# ============================================================
fig, ax = plt.subplots(figsize=(6,4))
bp = ax.boxplot([d_arr, c_arr], labels=['Drought recovery', 'Non-drought control'],
                patch_artist=True, showfliers=False)
bp['boxes'][0].set_facecolor('lightcoral')
bp['boxes'][1].set_facecolor('lightblue')
ax.set_ylabel('Absolute decoupling (|asymmetry|)')
ax.set_title('GPP–ET decoupling after drought vs. baseline')
ax.text(0.5, 0.95, f'p = {p_val:.5f}', transform=ax.transAxes, ha='center')
plt.tight_layout()
plt.savefig(os.path.join(OUTPUT_DIR, "decoupling_comparison.png"), dpi=150)
plt.show()
print(f"Plot saved to {OUTPUT_DIR}\\decoupling_comparison.png")
```

```python
"""
FLUXNET Extraction & Analysis - Jupyter Notebook
Fix: Timestamp parsing for multiple formats
"""

# ============================================================================
# CELL 1: IMPORTS & CONFIGURATION
# ============================================================================

import os
import sys
import zipfile
import tempfile
import shutil
import warnings
from pathlib import Path
from typing import Optional, Tuple
import numpy as np
import pandas as pd
from scipy.stats import mannwhitneyu
from datetime import datetime
import chardet
import matplotlib.pyplot as plt

warnings.filterwarnings('ignore')

# Configuration
DATA_PATH = r"C:\Users\BRAHMA LAB\Downlaod\mpi_fail\fluxnet"
OUTPUT_DIR = r"C:\Users\BRAHMA LAB\Downlaod\mpi_fail\fluxnet\analysis_output"
os.makedirs(OUTPUT_DIR, exist_ok=True)

# Constants
LE_TO_ET_FACTOR = 0.0353
VPD_PERCENTILE = 90
DROUGHT_MIN_DAYS = 5
DROUGHT_MERGE_DAYS = 30
PRE_DROUGHT_DAYS = 30
POST_DROUGHT_DAYS = 90
QC_THRESHOLD = 0.75
N_CONTROLS = 100
MIN_FLUX_GPP = 0.5
MIN_FLUX_ET = 0.2
DECOUPLING_CAP = 5.0

print("Configuration loaded ✓")

# ============================================================================
# CELL 2: DEBUG - CHECK TIMESTAMP FORMAT IN ONE FILE
# ============================================================================

def inspect_timestamp_format(zip_path: str):
    """Debug: Print actual timestamp format from first ZIP file."""
    temp_dir = tempfile.mkdtemp(dir=OUTPUT_DIR)
    try:
        with zipfile.ZipFile(zip_path, 'r') as z:
            z.extractall(temp_dir)
        
        # Find CSV
        for dirpath, dirnames, filenames in os.walk(temp_dir):
            for filename in filenames:
                if filename.lower().endswith('.csv') and 'fluxmet' in filename.lower():
                    csv_path = os.path.join(dirpath, filename)
                    df = pd.read_csv(csv_path, encoding='utf-8', nrows=5)
                    
                    print(f"File: {filename}")
                    print(f"Columns: {df.columns.tolist()}")
                    print("\nTimestamp column samples:")
                    
                    # Check for TIMESTAMP_START column
                    if 'TIMESTAMP_START' in df.columns:
                        print(f"TIMESTAMP_START values (first 5):")
                        print(df['TIMESTAMP_START'].head().tolist())
                        print(f"Data type: {df['TIMESTAMP_START'].dtype}")
                        print(f"Sample value: {df['TIMESTAMP_START'].iloc[0]}")
                        print(f"Length of first value: {len(str(df['TIMESTAMP_START'].iloc[0]))}")
                    
                    # Also check for other timestamp-like columns
                    ts_cols = [col for col in df.columns if 'TIME' in col.upper() or 'DATE' in col.upper()]
                    print(f"\nOther potential timestamp columns: {ts_cols}")
                    
                    return
    finally:
        shutil.rmtree(temp_dir)

# Inspect first ZIP file
zip_files = sorted([f for f in os.listdir(DATA_PATH) if f.endswith('.zip')])
if zip_files:
    print(f"Inspecting first ZIP: {zip_files[0]}\n")
    inspect_timestamp_format(os.path.join(DATA_PATH, zip_files[0]))

# ============================================================================
# CELL 3: IMPROVED TIMESTAMP PARSING (HANDLES MULTIPLE FORMATS)
# ============================================================================

def parse_timestamp_robust(ts_str) -> Optional[pd.Timestamp]:
    """
    Parse timestamp with fallback for multiple formats.
    Tries: YYYYMMDDHHMM, YYYYMMDDHH, YYYY-MM-DD HH:MM:SS, ISO format, etc.
    """
    if pd.isna(ts_str):
        return None
    
    ts_str = str(ts_str).strip()
    if not ts_str:
        return None
    
    # Formats to try
    formats = [
        '%Y%m%d%H%M',       # 12 digits: 202401010000
        '%Y%m%d%H',         # 10 digits: 2024010100
        '%Y-%m-%d %H:%M:%S',# ISO: 2024-01-01 00:00:00
        '%Y%m%d',           # 8 digits: 20240101
        '%Y-%m-%d',         # ISO date: 2024-01-01
    ]
    
    for fmt in formats:
        try:
            return pd.to_datetime(ts_str, format=fmt)
        except (ValueError, TypeError):
            continue
    
    # Last resort: pandas automatic parsing
    try:
        return pd.to_datetime(ts_str)
    except:
        return None

# Test the function
test_values = [
    '202401010000',      # YYYYMMDDHHMM
    '2024010100',        # YYYYMMDDHH
    '2024-01-01 00:00:00',# ISO
    '20240101',          # YYYYMMDD
]

print("Testing timestamp parsing:")
for val in test_values:
    result = parse_timestamp_robust(val)
    print(f"  {val} → {result}")

# ============================================================================
# CELL 4: HELPER FUNCTIONS
# ============================================================================

def read_csv_robust(filepath: str) -> Tuple[Optional[pd.DataFrame], str]:
    """Read CSV with encoding fallback."""
    encodings = ['utf-8', 'latin1', 'iso-8859-1', 'cp1252', 'windows-1252']
    
    for enc in encodings:
        try:
            df = pd.read_csv(filepath, encoding=enc, low_memory=False)
            return df, enc
        except (UnicodeDecodeError, UnicodeError, LookupError):
            continue
    return None, None

def find_column(df: pd.DataFrame, candidates):
    """Find first available column from candidates."""
    if isinstance(candidates, str):
        candidates = [candidates]
    for col in candidates:
        if col in df.columns:
            return col
    return None

def find_fluxmet_csv(root_dir: str) -> Optional[str]:
    """Find FLUXMET CSV in directory tree."""
    for dirpath, dirnames, filenames in os.walk(root_dir):
        for filename in filenames:
            if not filename.lower().endswith('.csv'):
                continue
            name_lower = filename.lower()
            has_fluxmet = 'fluxmet' in name_lower
            has_freq = '_hh_' in name_lower or '_hr_' in name_lower
            has_exclusion = 'bif' in name_lower or 'varinfo' in name_lower
            
            if has_fluxmet and has_freq and not has_exclusion:
                return os.path.join(dirpath, filename)
    return None

print("Helper functions loaded ✓")

# ============================================================================
# CELL 5: CORRECTED EXTRACTION FUNCTION
# ============================================================================

def extract_site_data(zip_path: str) -> Optional[pd.DataFrame]:
    """Extract FLUXNET FLUXMET data from ZIP."""
    temp_dir = None
    site_id = None
    
    try:
        print(f"Processing {os.path.basename(zip_path)}...", end=" ")
        
        # Stage 1: Create temp directory
        temp_dir = tempfile.mkdtemp(dir=OUTPUT_DIR, prefix='fluxnet_temp_')
        
        # Stage 2: Extract ZIP
        with zipfile.ZipFile(zip_path, 'r') as z:
            z.extractall(temp_dir)
        
        # Stage 3: Find CSV
        csv_path = find_fluxmet_csv(temp_dir)
        if not csv_path:
            print("[ERROR] No FLUXMET CSV found")
            return None
        
        # Stage 4: Read CSV
        df, encoding = read_csv_robust(csv_path)
        if df is None or df.empty:
            print("[ERROR] Failed to read CSV")
            return None
        
        print(f"Loaded {len(df)} rows", end=" → ")
        
        # Stage 5: Parse timestamp with robust parsing
        ts_col = find_column(df, 'TIMESTAMP_START')
        if not ts_col:
            print("[ERROR] No timestamp column")
            return None
        
        # Apply robust timestamp parsing
        df['datetime'] = df[ts_col].apply(parse_timestamp_robust)
        df = df.dropna(subset=['datetime'])
        
        if df.empty:
            print("[ERROR] No valid timestamps")
            return None
        
        # Extract site ID
        filename = os.path.basename(csv_path)
        parts = filename.split('_')
        site_id = parts[1] if len(parts) >= 2 else 'UNKNOWN'
        
        # Stage 6: Find columns
        gpp_col = find_column(df, ['GPP_NT_VUT_REF', 'GPP_DT_VUT_REF', 'GPP'])
        le_col = find_column(df, ['LE_F_MDS', 'LE_F', 'LE'])
        vpd_col = find_column(df, ['VPD_F_MDS', 'VPD_F', 'VPD'])
        
        if not gpp_col or not le_col:
            print("[ERROR] Missing GPP or LE column")
            return None
        
        # Stage 7: Apply QC filters
        qc_nee_col = find_column(df, ['NEE_VUT_REF_QC', 'NEE_QC'])
        qc_le_col = find_column(df, ['LE_F_MDS_QC', 'LE_QC'])
        
        if qc_nee_col:
            df = df[pd.to_numeric(df[qc_nee_col], errors='coerce') < QC_THRESHOLD]
        if qc_le_col:
            df = df[pd.to_numeric(df[qc_le_col], errors='coerce') < QC_THRESHOLD]
        
        if df.empty:
            print("[ERROR] All rows filtered by QC")
            return None
        
        # Stage 8: Convert to numeric & aggregate to daily
        df[gpp_col] = pd.to_numeric(df[gpp_col], errors='coerce')
        df[le_col] = pd.to_numeric(df[le_col], errors='coerce')
        if vpd_col:
            df[vpd_col] = pd.to_numeric(df[vpd_col], errors='coerce')
        
        df.set_index('datetime', inplace=True)
        df.sort_index(inplace=True)
        
        df_daily = pd.DataFrame()
        df_daily['gpp'] = df[gpp_col].resample('D').sum()
        df_daily['le'] = df[le_col].resample('D').mean()
        df_daily['et_mm'] = df_daily['le'] * LE_TO_ET_FACTOR
        
        if vpd_col:
            df_daily['vpd'] = df[vpd_col].resample('D').mean()
        else:
            df_daily['vpd'] = np.nan
        
        df_daily = df_daily.reset_index()
        df_daily['site'] = site_id
        df_daily = df_daily.dropna(subset=['gpp', 'et_mm'])
        
        if df_daily.empty:
            print("[ERROR] No daily data")
            return None
        
        df_daily = df_daily[['datetime', 'site', 'gpp', 'et_mm', 'vpd']]
        
        print(f"✓ {len(df_daily)} days ({site_id})")
        return df_daily
    
    except Exception as e:
        print(f"[ERROR] {str(e)}")
        return None
    
    finally:
        if temp_dir and os.path.exists(temp_dir):
            try:
                shutil.rmtree(temp_dir)
            except:
                pass

print("Extraction function ready ✓")

# ============================================================================
# CELL 6: LOAD ALL SITES
# ============================================================================

site_data = {}
zip_files = sorted([f for f in os.listdir(DATA_PATH) if f.endswith('.zip')])

print(f"\nLoading {len(zip_files)} sites...\n")

for zip_file in zip_files:
    zip_path = os.path.join(DATA_PATH, zip_file)
    df = extract_site_data(zip_path)
    
    if df is not None and not df.empty:
        site_id = df['site'].iloc[0]
        site_data[site_id] = df

print(f"\n✓ Successfully loaded {len(site_data)} sites\n")

# ============================================================================
# CELL 7: DROUGHT DETECTION & DECOUPLING ANALYSIS
# ============================================================================

def detect_droughts(df: pd.DataFrame, vpd_col: str = 'vpd') -> list:
    """Detect drought events using VPD threshold."""
    if vpd_col not in df.columns or df[vpd_col].isna().all():
        return []
    
    vpd_threshold = df[vpd_col].quantile(VPD_PERCENTILE / 100.0)
    drought_mask = df[vpd_col] > vpd_threshold
    
    drought_transitions = np.diff(drought_mask.astype(int))
    starts = np.where(drought_transitions == 1)[0] + 1
    ends = np.where(drought_transitions == -1)[0]
    
    if len(drought_mask) > 0 and drought_mask.iloc[0]:
        starts = np.insert(starts, 0, 0)
    if len(drought_mask) > 0 and drought_mask.iloc[-1]:
        ends = np.append(ends, len(df) - 1)
    
    events = [(int(s), int(e)) for s, e in zip(starts, ends) if (e - s + 1) >= DROUGHT_MIN_DAYS]
    
    if not events:
        return []
    
    merged = [events[0]]
    for s, e in events[1:]:
        prev_s, prev_e = merged[-1]
        if s - prev_e <= DROUGHT_MERGE_DAYS:
            merged[-1] = (prev_s, e)
        else:
            merged.append((s, e))
    
    return merged

def extract_window(df: pd.DataFrame, start_idx: int, end_idx: int) -> Optional[dict]:
    """Extract window statistics."""
    if start_idx < 0 or end_idx >= len(df):
        return None
    window = df.iloc[start_idx:end_idx+1]
    if window.empty:
        return None
    return {'gpp': window['gpp'].mean(), 'et': window['et_mm'].mean(), 'n': len(window)}

def compute_decoupling(pre_stats: dict, post_stats: dict) -> Optional[float]:
    """Compute decoupling metric."""
    if not pre_stats or not post_stats:
        return None
    gpp_pre, et_pre = pre_stats['gpp'], pre_stats['et']
    gpp_post, et_post = post_stats['gpp'], post_stats['et']
    
    if gpp_pre < MIN_FLUX_GPP or et_pre < MIN_FLUX_ET or gpp_pre == 0 or et_pre == 0:
        return None
    
    et_ratio = et_post / et_pre
    gpp_ratio = gpp_post / gpp_pre
    asym = et_ratio - gpp_ratio
    decoupling = abs(asym)
    
    return min(decoupling, DECOUPLING_CAP)

def analyze_droughts(df_site: pd.DataFrame) -> Tuple[list, list]:
    """Analyze droughts at a site."""
    droughts = detect_droughts(df_site)
    if not droughts:
        return [], []
    
    decouplings = []
    for s, e in droughts:
        pre_start = s - PRE_DROUGHT_DAYS
        post_end = e + POST_DROUGHT_DAYS
        
        if pre_start < 0 or post_end >= len(df_site):
            continue
        
        pre_stats = extract_window(df_site, pre_start, s - 1)
        post_stats = extract_window(df_site, e + 1, post_end)
        
        decoup = compute_decoupling(pre_stats, post_stats)
        if decoup is not None:
            decouplings.append(decoup)
    
    return decouplings, [0] * len(decouplings)

# Run analysis
print("Analyzing droughts...\n")

results = []
drought_means = []
control_means = []

for site_id in sorted(site_data.keys()):
    df_site = site_data[site_id].sort_values('datetime').reset_index(drop=True)
    
    droughts = detect_droughts(df_site)
    drought_decouplings, _ = analyze_droughts(df_site)
    
    # Generate control windows
    window_length = PRE_DROUGHT_DAYS + POST_DROUGHT_DAYS
    drought_mask = np.zeros(len(df_site), dtype=bool)
    for s, e in droughts:
        drought_mask[max(0, s - PRE_DROUGHT_DAYS) : min(len(df_site), e + POST_DROUGHT_DAYS)] = True
    
    valid_starts = [i for i in range(len(df_site) - window_length) if not drought_mask[i:i+window_length].any()]
    
    controls = []
    if valid_starts:
        np.random.seed(42)
        for _ in range(min(N_CONTROLS, len(valid_starts))):
            start_idx = np.random.choice(valid_starts)
            pre_stats = extract_window(df_site, start_idx, start_idx + PRE_DROUGHT_DAYS - 1)
            post_stats = extract_window(df_site, start_idx + PRE_DROUGHT_DAYS, start_idx + window_length - 1)
            decoup = compute_decoupling(pre_stats, post_stats)
            if decoup is not None:
                controls.append(decoup)
    
    drought_mean = np.mean(drought_decouplings) if drought_decouplings else np.nan
    control_mean = np.mean(controls) if controls else np.nan
    
    results.append({
        'site': site_id,
        'n_droughts': len(droughts),
        'drought_decoupling_mean': drought_mean,
        'control_decoupling_mean': control_mean,
    })
    
    if np.isfinite(drought_mean):
        drought_means.append(drought_mean)
    if np.isfinite(control_mean):
        control_means.append(control_mean)
    
    print(f"{site_id:12s}: {len(droughts):2d} droughts, drought={drought_mean:.3f}, control={control_mean:.3f}")

print()

# ============================================================================
# CELL 8: STATISTICAL TESTING & VISUALIZATION
# ============================================================================

drought_means = np.array(drought_means)
control_means = np.array(control_means)

print(f"Sites with droughts: {len(drought_means)}")
print(f"Drought mean: {drought_means.mean():.4f} ± {drought_means.std():.4f}")
print(f"Control mean: {control_means.mean():.4f} ± {control_means.std():.4f}")

# Mann-Whitney U test
stat, pval = mannwhitneyu(drought_means, control_means, alternative='greater')
print(f"\nMann-Whitney U Test:")
print(f"  p-value: {pval:.6f}")

# Bootstrap CI
np.random.seed(42)
bootstrap_diffs = []
for _ in range(10000):
    d_sample = np.random.choice(drought_means, size=len(drought_means), replace=True)
    c_sample = np.random.choice(control_means, size=len(control_means), replace=True)
    bootstrap_diffs.append(d_sample.mean() - c_sample.mean())

ci_lower = np.quantile(bootstrap_diffs, 0.025)
ci_upper = np.quantile(bootstrap_diffs, 0.975)
print(f"  95% CI: [{ci_lower:.4f}, {ci_upper:.4f}]")

# Plot
fig, ax = plt.subplots(figsize=(10, 6))
bp = ax.boxplot([drought_means, control_means], labels=['Drought', 'Control'], patch_artist=True)

for patch, color in zip(bp['boxes'], ['#ff7f0e', '#1f77b4']):
    patch.set_facecolor(color)
    patch.set_alpha(0.7)

ax.set_ylabel('Decoupling |asym|', fontweight='bold')
ax.set_title('Drought-Induced GPP-ET Decoupling', fontweight='bold')
ax.grid(True, alpha=0.3)

y_max = max(np.max(drought_means), np.max(control_means))
ax.text(1.5, y_max * 1.1, f'p = {pval:.4f}', ha='center', fontsize=11, fontweight='bold',
        bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.8))

plt.tight_layout()
plt.savefig(os.path.join(OUTPUT_DIR, 'decoupling_boxplot.png'), dpi=300)
plt.show()

print("\n✓ Analysis complete!")
```
